```python
#2) Write a function to calculate the mean of a specific column in a DataFrame
def calculate_column_mean(df, column_name):

    column_mean = df[column_name].mean()
    return column_mean
  
calculate_column_mean(df, 2)
```

```python
#3) How do you handle missing values in a DataFrame? Use a CSV and handle missing values in it?

df1 = pd.read_csv("ckd.csv")
def handle_missing_values(df):
    null_counts = df.isnull().sum()
    columns_to_impute = null_counts[null_counts < 20].index
    
    for column in columns_to_impute:
        if df[column].dtype == 'object':
            # Categorical column: fill with mode
            df[column] = df[column].fillna(df[column].mode().iloc[0])
        else:
            # Numerical column: fill with mean
            df[column] = df[column].fillna(df[column].mean())
    
    # Drop rows with any null values
    df = df.dropna()
    
    
    return df

cleaned_df = handle_missing_values(df1)
cleaned_df
```

```python
#8) Write a function to convert the index of a DataFrame into a column.

def convert_index_to_column(df):

    df_with_index_as_column = df.reset_index()
    return df_with_index_as_column
  
converted_df = convert_index_to_column(df1)
converted_df
```
